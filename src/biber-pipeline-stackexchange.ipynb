## SE Culture

```python
dfse_cul = pd.read_csv(stex_cul)
dfse_cul
```

```python
print(dfse_cul.isna().sum().to_frame(name="Null Count").to_string())
```

```python
assert dfse_cul['content_id_domain'].drop_duplicates().shape[0] == dfse_cul.shape[0]
dfse_cul.rename(columns={'content_id_domain':'documentID', 'user_id_domain':'authorID', 'word_count':'fullText_space_lengthWords', 'message':'fullText'}, inplace=True)
```

```python
min_author_docs_count = 10
min_batch_authors_sample = 2000
batch_biber_ = 1
clip_authors_docs_count = 15
```

```python
df_gte70 = dfse_cul.loc[dfse_cul['fullText_space_lengthWords'] >= 70]
df_gte70['author_docs_count'] = df_gte70.groupby('authorID')['documentID'].transform('count')
df_gte70_author_docs_filtered = df_gte70.loc[df_gte70['author_docs_count'] >= min_author_docs_count]
print(f"\tFiltered Shape: {df_gte70_author_docs_filtered.shape}")

df_gte70_uniq_authors = df_gte70_author_docs_filtered.drop_duplicates(subset='authorID', keep='first')[['authorID']]
print(f"\tUnique Authors Shape: {df_gte70_uniq_authors.shape}")

df_gte70_uniq_authors_shuffled = shuffle(df_gte70_uniq_authors, random_state=42).reset_index(drop=True)

# Compute start and end index for slicing
start_index = (batch_biber_ - 1) * min_batch_authors_sample
end_index = start_index + min_batch_authors_sample  # Fetch exactly `min_batch_authors_sample` rows
df_gte70_uniq_authors_sample = df_gte70_uniq_authors_shuffled.iloc[start_index:end_index]
print(f"\tAuthors Sample Shape: {df_gte70_uniq_authors_sample.shape}")

df_gte70_author_docs_filtered_batch_sample = df_gte70_author_docs_filtered.merge(df_gte70_uniq_authors_sample, on='authorID')
print(f"\tMerged Shape: {df_gte70_author_docs_filtered_batch_sample.shape}")

df_gte70_author_docs_filtered_batch_sample = df_gte70_author_docs_filtered_batch_sample.groupby('authorID', group_keys=False).apply(
    lambda group: group.sample(n=clip_authors_docs_count, random_state=42) if len(group) > clip_authors_docs_count else group
)

df_gte70_author_docs_filtered_batch_sample["source_of_biber"] = "stex_" + df_gte70_author_docs_filtered_batch_sample['category'].str.lower()
df_gte70_author_docs_filtered_batch_sample["min_length_words_biber"] = "gte70"
df_gte70_author_docs_filtered_batch_sample["batch_biber"] = batch_biber_
df_gte70_author_docs_filtered_batch_sample["min_author_docs_count_biber"] = min_author_docs_count
df_gte70_author_docs_filtered_batch_sample["min_batch_authors_sample_biber"] = min_batch_authors_sample

print(f"Final DF Shape: {df_gte70_author_docs_filtered_batch_sample.shape}")
df_gte70_author_docs_filtered_batch_sample.to_json(stex_op_cul_gte70, orient='records', lines=True)
```
