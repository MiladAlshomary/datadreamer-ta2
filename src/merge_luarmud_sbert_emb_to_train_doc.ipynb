```python
import os
dir = "/data/araghavan/HIATUS/datadreamer-ta2/data/ta2_jan_2025_trian_data/trainsadiri_luarmud_chunks/"
dirfiles = os.listdir(dir)
dirfiles_luarembds_only = [os.path.join(dir,x) for x in dirfiles if "train_sadiri_processed_with_luarembeddings" in x]
dirfiles_luarembds_only.sort()
dirfiles_luarembds_only
```

```python
import pandas as pd
df_arr = []
for file in dirfiles_luarembds_only:
    print(f"Processing File: {file}")
    df = pd.read_json(file, lines=True)
    df_arr.append(df)
```

```python
df_luaremb = pd.concat(df_arr, ignore_index=True)
```

```python
file_sbertemb = "/data/araghavan/HIATUS/datadreamer-ta2/data/ta2_jan_2025_trian_data/train_sadiri_processed_with_embeddings_wo_ao3_filtered.jsonl"
df_sbertemb = pd.read_json(file_sbertemb, lines=True)
print(f"Shape of SBERT Embedding File: {df_sbertemb.shape}")
```

```python
df_luarsbert_emb = df_sbertemb.merge(df_luaremb[['documentID', 'doc_luarmud_embedding', 'doc_xrbmtgc_genre']], how='inner', on='documentID')
df_luarsbert_emb.rename(columns={'embeddings':'doc_sbertamllv2_embedding'}, inplace=True)
```
